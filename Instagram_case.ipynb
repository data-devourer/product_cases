## Задача:

Представьте себя в роли продуктового аналитика в Instagram, ответственного за анализ функционала Stories. Прошло несколько месяцев с момента его запуска, и вас попросили создать продуктовый дашборд, на котором бы отражались результаты успешности запуска этой функции, начальные показатели работы продукта и его воздействие на общую производительность Instagram. Опишите метрики, которые вы бы включили в этот дашборд, упорядоченные по убыванию их важности.

## Знакомство с продуктом
____Instagram Stories____ - это короткие фотографии или видеоролики, которыми пользователь _Instagram_ может поделиться с подписчиками в течение 24 часов. Stories являются временными публикациями и исчезают автоматически через промежуток времени.

> Истории позволяют пользователям делиться моментами своей жизни, впечатлениями, фотографиями и видео в более непринужденной форме, чем обычные публикации в ленте. Как правило, они имеют более неформальный характер и могут содержать текст, эмодзи, рисунки и другие дополнительные элементы.

---

## Особенности продукта
Для удобства выделим основные характеристики продукта:

- __Время существования__: Истории существуют только 24 часа после публикации, после чего они исчезают
- __Разнообразие форматов__: Пользователи могут использовать фотографии, видео, текст, эмодзи, стикеры, рисунки и другие элементы для создания своих сторис
- __Приватность__: Пользователи могут опубликовать истории для всех своих подписчиков или ограничить доступ только для определенных пользователей
- __Интерактивность__: Сторис позволяют взаимодействовать с подписчиками с помощью опросов, вопросов, голосований и других интерактивных элементов.
- __История просмотров__: Автор истории может увидеть, кто просмотрел каждую конкретную историю.

> Stories в Instagram стали популярным способом делиться мгновенными впечатлениями и событиями. Stories имеют широкое применение среди как обычных пользователей, так и брендов.

## Пулл метрик
Для оценки успешности продукта сторис в _Instagram_ можно использовать различные метрики, которые предоставят возможности измерить вовлеченность аудитории, эффективность контента и общий влияние сторис на аудиторию. 
Условно поделим подходящие метрики на группы по видам:

### __Использование продукта__:

- Доля активных пользователей, которые создают истории в Instagram
- DAU, WAU, MAU (пользователи историй, которые создают или просматривают хотя бы n историй до конца за соответствующий период времени)
  - Количество созданных историй (в день, в неделю, в месяц)
  - Количесво просмотренных историй (в день, в неделю, в месяц)
- Частота использования сторис (среднее количество созданных/просмотренных сторис на пользователя за определенный период)
- Среднее количество созданных сторис на каждого активного пользователя

### __Вовлеченность__:

- Среднее время, проведенное пользователем на взаимодействие с историями
- Среднее время, за одну активную сессию на взаимодействие с историями
- % пользователей, взаимодействующих с историями посредством комментариев/лайков/репостов
- % пользователей, досмотревших историю до конца
- % пользователей, тратящих более 10 минут на скроллинг сторис за сессию

### __Retention (удержание)__:

- Значение retention на 1, 3, 7, 14 день для пользователей, взаимодействующих с историями Instagram
  - Значение retention на 1, 3, 7, 14 день для создателей сторис
  - Значение retention на 1, 3, 7, 14 день для потребителей сторис
- Churn Rate (процент пользователей, которые перестали использовать продукт после определенного периода времени)
- % трафика, переходящего по ссылкам в сторис ежедневно

### __Монетизация__:

- ARPU (средний доход, полученный от одного пользователя за определенный период)
- LTV (прогнозируемая сумма дохода от одного пользователя за всё время его использования продукта)
- % дохода от рекламы в сторис от всего дохода продукта Instagram
- % рекламных размещений, которые сделаны в сторис по сравнению с общим числом размещений в Instagram
- Среднее количество показов рекламы в сторис
- Среднее количество кликов на рекламу в сторис

## Непосредственно топ

1. ____Консолидированная метрика____ ($CM = n_1 * AU + n_2 * AT + n_3 * Ret + n_4 * AdI$, где $n_i$ - коэффициенты нормировки) - позволяет учесть различные аспекты; преимущество метрики в том, что она дает полное представление о состоянии продукта с разных сторон: от вовлеченности пользователей до успешности продукта с точки зрения доходов для Meta
2. ____Доля активных пользователей, которые создают/просматривают истории в Instagram____ ($AUP$) - показывает, насколько успешно пользователи включаются в создание/потребление контента. Высокий процент указывает на активное сообщество и интерес пользователей к общению и обмену информащии в формате Stories
3. ____Среднее время, проведенное пользователем на взаимодействие с историями____ ($AT$) - отражает уровень вовлеченности пользователей к продукту. Большее время, проведенное на взаимодействие, указывает на приятный для пользователя опыт взаимодействия с продуктом, качественный и интуитивно понятный интерфейс
4. ____Значение Retention на 1, 3, 7, 14 день____ ($Ret$) - позволяет измерить, насколько успешно Stories удерживает как создателей, так и потребителей. Более высокий уровень удержания указывает на вовлеченность уже на более значительных временных промежутках, на потребность аудитории в ежедневном общении/самовыражении (способность Stories удовлетворять этой потребности)
5. ____% дохода от рекламы в историях____ ($AdI$) - имеет большое значение для понимания, насколько успешно контент в Stories привлекает рекламодателей. Это непосредственно отражает, насколько продукт успешен с точки зрения бизнеса. Метрика позволяет оценить не только абсолютный рост стоимости продукта, но и увеличение доли дохода внутри всей платформы Instagram
